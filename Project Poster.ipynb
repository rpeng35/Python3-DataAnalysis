### Determine the most number of songs in the list in the year range 1990 - 2010

We are given a massive list of songs and each song has a title, artist, year, popularity, duration and tempo. The list of songs is from the sample information scource on canvas called songs.csv. For this project, we are trying to figure out the number of songs in each year in the year range of 1990 to 2010. We want to know how many songs there are in each year in the 21-year period and figure out which year has the most songs.

Systematic program design keeps me focused on the elements I need to work with in a messy list. By following the steps, it is easy to spot an error and fix it right away. This keeps me on track and guides me in reaching my final funtion which returns the value I want.
Some of the design choices I made were:
 - filtering the years (in to the range I am interested in) before putting the data in my analyze function
 - using one single accumulator with 21 slots (List[int]) instead of 21 seperate accumulators for each year

My partner and I decided to use a bar graph to help determine the problem that we are trying to solve which is to determine the year with the most number of songs. Here is the graph that our program produced. We figured that a bar graph would make the most sense in this problem.


The most difficult part of the project was to create the function that produces my y-values which is the number of songs (List[int]). I had to figure out how to set up my function so it only uses one accumulator instead of 21 seperate accumulators for each year. If I chose to create 21 accumulators for each year, it would take around 100 lines of code to complete plus an additional helper function would be required, whereas one single accumulator with 21 slots is much shorter and more efficient to use in this case.

def filter_songs(los: List[Songs]) -> List[int]:
    '''
    takes a list of songs and returns a list of integer indicating the number of songs that corresponds
    to each year
    '''
    #return [] #stub
    #template based on List[Songs]
    
    #acc description
    acc = [0] * 21 # type: List[int]
    for s in los:
            acc[s.year - 1990] = acc[s.year - 1990] + 1

    return acc
